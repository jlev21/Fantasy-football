```python
df_teams.rename(columns={'PLAYER NAME': 'Player'}, inplace=True)
df_teams.rename(columns={'TEAM': 'Team'}, inplace=True)
```

```python
df = pd.merge(df, df_teams, on='Player', how='inner')
df.rename(columns={'Team_y': 'curr_team'}, inplace=True)
df.rename(columns={'Team_x': 'old_team'}, inplace=True)
```

```python
df = df.drop_duplicates(subset='Player', keep='last')
```

```python
df.head(20)
```

```python
new_to_old_map = {
    'LV': 'LVR',
    'SF': 'SFO',
    'TB': 'TAM',
    'KC': 'KAN',
    'GB': 'GNB',
    'NE': 'NWE',
    'NO': 'NOR',
    'JAC': 'JAX',
    'FA': 'ATL' 
}

def map_new_to_old(abbr):
    return new_to_old_map.get(abbr, abbr)

# Example usage with a DataFrame:
df['curr_team'] = df['curr_team'].apply(map_new_to_old)
```

```python
df['curr_team'].unique()
```

```python
sos_df = pd.read_csv('2025_SOS.csv')
```

```python
sos_df.rename(columns={'Team': 'curr_team'}, inplace=True)
sos_df.rename(columns={'Win_Percentage': 'SOS'}, inplace=True)
sos_df.head(20)
```

```python
df2 = pd.merge(df, sos_df[['curr_team','SOS']], on="curr_team", how="inner")
```

```python
df3 = df2[df2['Pos']=='WR']
```

```python
df_encoded = pd.get_dummies(df3, columns=['old_team', 'curr_team'], drop_first=True)
```

```python
df_encoded = df_encoded.drop(columns=['Awards','Pos','Rec/Br'])
```

```python
df4 = df_encoded.dropna()
```

```python
fantasy = pd.read_csv('fantasy.csv')
```

```python
df_merged = pd.merge(df4, fantasy[['Player', 'AVG']], on='Player', how='inner')
df_merged.rename(columns={'AVG': 'avg_old'}, inplace=True)
```

```python
df5 = df_merged.drop(columns=['Player']).astype(float)
```

```python
df5.head()
```

```python
import joblib
loaded_svr_model = joblib.load('svr.pkl')
```

```python
# Make predictions on the new data using the loaded model
new_predictions = loaded_svr_model.predict(df5)
df5['predictions'] = new_predictions
df5
```

```python
df5.sort_values(by="predictions", ascending=False, inplace=True)
df5_merged = df5.merge(df[['old_rank', 'Player']], on='old_rank', how='inner')
```
